```python
pcpop
```

```python
pcpop.drop('SteamPopulation',axis=1,inplace=True)
```

```python
tes = pd.DataFrame([[17955645,17955645]],['SteamPopulation'],columns=['PeakPlayers','AvgConcurrent'])
```

```python
frames = [tes,pcpop]
```

```python
result = pd.concat(frames)
```

```python
result
```

```python
result['ConcurrentToPeak(%)'] = result['AvgConcurrent'].apply(lambda x: x/15228924*100)
```

```python
result
```

```python
result['$50'] = result['PeakPlayers'].apply(lambda x: x*50)
result['$25'] = result['PeakPlayers'].apply(lambda x: x*25)
result['$10'] = result['PeakPlayers'].apply(lambda x: x*10)
result
```

```python
def steamcut(x):
    if x < 10000000:
        return x - (x * 0.3)
    if x > 10000000 and x < 50000000:
        return x - 3000000 - (x * 0.25)
    if x > 50000000:
        return x - 13000000 - (x * 0.2)
```

```python
result['$50SteamCut'] = result['$50'].apply(steamcut)
result['$25SteamCut'] = result['$25'].apply(steamcut)
result['$10SteamCut'] = result['$10'].apply(steamcut)
result
```

```python
result.reset_index(inplace=True)
```

```python
result
```

```python
result[['index','PeakPlayers','AvgConcurrent','$50','$50SteamCut','$25','$25SteamCut','$10','$10SteamCut']]
```

### Android Premium

```python
android = pd.DataFrame(['Minecraft','Monopoly','Stardew Valley','Bloons TD6','Grand Theft Auto: San Andreas','Exploding Kittens','The Game of Life','Geometry Dash','Five Nights at Freddys','Terraria'],columns=['Name'])
```

```python
android['price'] = 6.99,3.99,7.99,4.99,6.99,1.99,2.99,1.99,2.99,4.99
android
```

```python
android['downloads(in thousands)'] = 10000,100,500,1000,1000,500,500,1000,1000,1000
```

```python
def pricing(cols):
    return cols['price'] * cols['downloads(in thousands)']
```

```python
android['earnings(in thousands)'] = android[['price','downloads(in thousands)']].apply(pricing, axis =1)
```

```python
android
```

```python
android.drop('price', axis = 1).plot(x = 'Name',kind='barh')
```
